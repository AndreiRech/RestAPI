# LIMPEZA DOS DADOS

Por conta da grande quantidade de informações que *não são utilizadas e nem relevantes* no contexto do trabalho, é necessário realizar uma **limpeza dos dados** antes de começar a utilizar o dataset propriamente.

```python
import pandas as pd
```

Primeiro, vizualizamos as informações contidas no dataset utilizando a biblioteca **pandas**

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('games.csv')
df.head()
```

Após isso, selecionamos as colunas **indesejadas** e removemos elas do dataset.

```python
colunas_droppar = ["AppID","Required age", "Score rank", "Notes", "User score","error","Peak CCU", "Supported languages", "Recommendations", "DiscountDLC count", "Full audio languages", "Reviews", "Header image", "Website", "Support url", "Support email", "Windows", "Mac", "Linux", "Metacritic url", "Screenshots", "Movies"]
```

```python
novo = df.drop(columns=colunas_droppar)

novo.head()
```

Num último momento, apenas finalizamos testando um caso e salvamos como um arquivo que será utilizado no restante do trabalho.

```python
print(novo[novo['Name'] == "Counter-Strike: Global Offensive"])
```
